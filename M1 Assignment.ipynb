```python
import pandas as pd

url = "https://raw.githubusercontent.com/cosc-526/home.page/main/data.M.1.assignment.covid.data.csv"
df = pd.read_csv(url)

category1_variants = ["Alpha", "Delta", "Omicron", "others", "non_who"]

US_and_other = df[df["variant"].isin(category1_variants)]["variant"].unique()

nonUS_and_other = df[~df["variant"].isin(category1_variants)]["variant"].unique()

print("US_and_other:", US_and_other)
print("nonUS_and_other:", nonUS_and_other)

print("US_and_other count:", len(US_and_other))
print("nonUS_and_other count:", len(nonUS_and_other))
print("Total unique variants:", df["variant"].nunique())


```

```python
variant_most_proc = df.groupby('variant')['num_sequences'].sum().idxmax()
print(variant_most_proc)
```

Task.3

```python
country_seq_stats = df.groupby('location').aggregate({
    'num_sequences': 'sum',
    'num_sequences_total': 'sum'
})

country_seq_stats['perc_sequences'] = (
    country_seq_stats['num_sequences'] / country_seq_stats['num_sequences_total']
) * 100

print(country_seq_stats.sort_values(by='perc_sequences', ascending=False)[['perc_sequences']])

best_proc_country = country_seq_stats['perc_sequences'].idxmax()
best_value = country_seq_stats['perc_sequences'].max()

print(f"The country with the highest percentage of processed sequences is {best_proc_country} with {best_value:.2f}%")
```

```python
ADO_variants = ["Alpha", "Delta", "Omicron"]
df_ADO = df[df["variant"].isin(ADO_variants)]

country_ADO_stats = df_ADO.groupby("location").agg({
    "num_sequences": "sum",
    "num_sequences_total": "sum"
})

country_ADO_stats["perc_sequences"] = (
    country_ADO_stats["num_sequences"] /
    country_ADO_stats["num_sequences_total"]
) * 100

best_proc_country_ADO = country_ADO_stats["perc_sequences"].idxmax()

print("best_proc_country_ADO:", best_proc_country_ADO)
```

```python

# 1. Filter for Alpha, Delta, Omicron
ADO_variants = ["Alpha", "Delta", "Omicron"]
df_ADO = df[df["variant"].isin(ADO_variants)]

# 2. Group by country and compute sums
country_ADO_stats = df_ADO.groupby("location").agg({
    "num_sequences": "sum",
    "num_sequences_total": "sum"
})

# 3. Compute processing percentage
country_ADO_stats["perc_sequences"] = (
    country_ADO_stats["num_sequences"] /
    country_ADO_stats["num_sequences_total"]
) * 100

# 4. Sort best → worst
sorted_ADO = country_ADO_stats.sort_values(by="perc_sequences", ascending=False)

# 5. Find U.S. ranking (adjust name if needed)
us_index = sorted_ADO.index.get_loc("United States")   # or "United States" depending on dataset

# 6. Convert 0‑based index → 1‑based ranking
us_ranking = us_index + 1

# 7. Print result
print(f"United States ranking: {us_ranking}")
```
