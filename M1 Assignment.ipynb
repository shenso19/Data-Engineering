Task.3

```python
country_seq_stats = df.groupby('location').aggregate({'num_sequences': 'sum', 'num_sequences_total': 'sum'})
country_seq_stats['perc_sequences'] = (country_seq_stats['num_sequences'] / country_seq_stats['num_sequences_total']) * 100
print(country_seq_stats.sort_values(by='perc_sequences', ascending=False)[['perc_sequences']])

best_proc_country = country_seq_stats['perc_sequences'].idxmax()
print(f"The country with the highest percentage of processed sequences is: {best_proc_country}")
```
